Initial efforts will focus on examining what factors may contribute to graduation rates. Graduate rates help determine whether schools are designated as Title I schools, and thus receive significant funding for special programs to aid students and teachers. 

```python
demodata[['School', 'Adjusted.5.Year.Cohort.Graduation.Rate']]
```

```python
import matplotlib.pyplot as plt
ax = demodata[['School','Adjusted.5.Year.Cohort.Graduation.Rate']].plot(kind='bar')
N = 3
params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches( (plSize[0]*N, plSize[1]*N) )
plt.show()
```

We can see in the plot above that there are a few schools with alarmingly low graduation rates. We will look at schools that have less than a 20% graduation rate.

```python
demodata['School'][demodata['Adjusted.5.Year.Cohort.Graduation.Rate']<20]
```

A few of these schools may just be special programs, which might not actually "graduate" students. This would explain a few of the high dropout rates. However, there are a fair number of regular schools on this list! 

```python
lowGradRateSchools = demodata['School'][demodata['Adjusted.5.Year.Cohort.Graduation.Rate']<20]
lowGradRateSchools_data = demodata[demodata['School'].isin(lowGradRateSchools)]
```

```python
import matplotlib.pyplot as plt
ax = lowGradRateSchools_data[['School','Adjusted.5.Year.Cohort.Graduation.Rate']].plot(kind='bar')
N = 3
params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches( (plSize[0]*N, plSize[1]*N) )
my_xticks = lowGradRateSchools
plt.xticks(range(0,len(lowGradRateSchools)), list(my_xticks))
plt.show()
```

We will examine this group of high dropout schools are programs further to determine what factors might be correlated with these low graduation rates. The following is a list of factors (continuous variables) that we will examine.

```python
varInt = ['NumberAmericanIndianorAlaskanNative',
          'PercentAmericanIndianorAlaskanNative',
          'NumberAsian', 'PercentAsian', 'NumberPacificIslander', 'PercentPacificIslander',
          'NumberAsianPacificIslander', 'PercentAsianPacificIslander', 'NumberBlack', 'PercentBlack', 
          'NumberHispanic', 'PercentHispanic',
          'NumberWhite', 'PercentWhite', 'NumberTwoOrMoreRaces', 
          'PercentTwoOrMoreRaces', 'NumberMales', 'PercentMales',
          'NumberFemales', 'PercentFemales', 'NumberMigrant',
          'PercentMigrant', 'NumberTransitionalBilingual',
          'PercentTransitionalBilingual', 'NumberSpecialEducation',
          'PercentSpecialEducation', 'NumberFreeorReducedPricedMeals',
          'PercentFreeorReducedPricedMeals', 'NumberSection504',
          'PercentSection504', 'PercentFosterCare',
          'StudentsPerClassroomTeacher', 'AvgYearsEducationalExperience', 'NumberTeachersWithAtLeastMasterDegree',
          'PercentTeachersWithAtLeastMasterDegree', 'Began.Grade.9.in.Washington',
          'Transferred.Into.Washington', 'Year.1.Dropout', 'Year.2.Dropout',
          'Year.3.Dropout', 'Year.4.Dropout', 'Year.5.Dropout',
          'Transferred.Out', 'Adjusted.Cohort', 'Graduates', 'Continuing',
          'Adjusted.5.Year.Cohort.Graduation.Rate', 'Cohort.dropout.rate',
          'Continuing.Rate']
```

Here we plot all factors of interest and their relationship to graduation rate.

```python
for i in range(0,len(varInt)):
    ax = demodata[[varInt[i],'Adjusted.5.Year.Cohort.Graduation.Rate']].plot.scatter(x=varInt[i], y='Adjusted.5.Year.Cohort.Graduation.Rate')
    N = 1.5
    params = plt.gcf()
    plSize = params.get_size_inches()
    params.set_size_inches( (plSize[0]*N, plSize[1]*N) )
    plt.show()
```

Interestingly, there appears to be a relationship between Percent Special Education and Graduation Rate. It appears that more schools with low graduation rates have very low percentages of special education students.

```python
import matplotlib.pyplot as plt
ax = demodata[['PercentSpecialEducation','Adjusted.5.Year.Cohort.Graduation.Rate']].plot.scatter(x='PercentSpecialEducation', y='Adjusted.5.Year.Cohort.Graduation.Rate')
N = 3
params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches( (plSize[0]*N, plSize[1]*N) )
plt.show()
```

```python
import matplotlib.pyplot as plt
ax = lowGradRateSchools_data[['PercentSpecialEducation','Adjusted.5.Year.Cohort.Graduation.Rate']].plot.scatter(x='PercentSpecialEducation', y='Adjusted.5.Year.Cohort.Graduation.Rate')
N = 3
params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches( (plSize[0]*N, plSize[1]*N) )
plt.show()
```

What could be significant about schools with low graduation rates having low percentages of special education students? It is possible that we can infer that these schools actually just have few students in special education because of a lack of resources, as opposed to just having few special education students. If this is the case, it is safe to say that any special education students are being under-served in these schools, which could indeed lead to poor graduation rates!

Further exploration and analysis of these schools of interest and their individual students can help clarify contributing factors.
